```python
import os
import matplotlib.pyplot as plt
from IPython.display import Video
from main import get_accident_risk_prediction
```

```python
# Set the video path
video_path = os.path.join(os.getcwd(), 'sample data', '000431.mp4')
print(f'Processing video: {video_path}')
```

```python
# Get predictions
predictions = get_accident_risk_prediction(video_path)
print(f'Predictions: {predictions}')
```

```python
# Plot the risk scores
plt.figure(figsize=(10, 5))
plt.plot(predictions, marker='o', linestyle='-', color='r')
plt.title('Accident Risk Score over Time')
plt.xlabel('Time Step (Sequence)')
plt.ylabel('Risk Score (0-1)')
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```
